```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
451languages = 'Python', 'Java', 'PHP', 'JavaScript', 'C#', 'C++'
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
```

```python
;mn colors = ['#0099cc', '#cc66ff', '#66ff66', '#ff6666', '#00cc99', '#c68c53']
```

```python
plt.pie(popularity, labels=languages, colors=colors)
plt.title('Popularity of Programming Languages in 2019')
plt.show()
```

```python
explode = (0.1, 0, 0.2, 0, 0.2, 0)
plt.pie(popularity, labels=languages, colors=colors, autopct='%1.1f%%', explode=explode, shadow=True,
        startangle=90, rotatelabels=True, wedgeprops = {'linewidth': 3}, radius=2)
plt.show()
```
