```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
values = [82,76,24,40,67,62,75,78,71,32,98,89,78,67,72,82,87,66,56,52]
values1 = [42,36,24,40,67,62,45,58,41,32,48,49,38,57,42,72,57,56,56,43]
```

```python
plt.hist(values, histtype='bar', edgecolor='black', label='test results')
plt.title('Histogram of Scores')
plt.ylabel('Frequency')
plt.xlabel('Scores')
plt.legend()
plt.show()
```

```python
plt.hist(values, histtype='bar', edgecolor='black', label='test results', range=[0,100])
plt.title('Histogram of Scores')
plt.ylabel('Frequency')
plt.xlabel('Scores')
plt.legend()
plt.show()
```

```python
plt.hist(values, histtype='bar', edgecolor='black', label='test results', range=[0,100], orientation='horizontal')
plt.title('Histogram of Scores')
plt.ylabel('Frequency')
plt.xlabel('Scores')
plt.legend()
plt.show()
```

```python
plt.hist(values, histtype='bar', edgecolor='black', label='old test results', range=[0,100])
plt.hist(values1, histtype='barstacked', edgecolor='black', label='new test results', range=[0,100])
plt.title('Histogram of Scores')
plt.ylabel('Frequency')
plt.xlabel('Scores')
plt.legend()
plt.show()
```

```python
plt.hist(values, histtype='bar', edgecolor='black', label='old test results', range=[0,100], alpha=0.5)
plt.hist(values1, histtype='barstacked', edgecolor='black', label='new test results', range=[0,100], alpha=0.3)
plt.title('Histogram of Scores')
plt.ylabel('Frequency')
plt.xlabel('Scores')
plt.legend()
plt.show()
```
